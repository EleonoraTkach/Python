```python
#1 Копирование содержимого одного файла в другой
try:
    source_file = open('source.txt','r',encoding='utf-8')
    content = source_file.readlines()
    source_file.close()
    
    descination_file = open('destination.txt','w',encoding='utf-8')
    for i in content:
        descination_file.write(i)
        
except FileNotFoundError:
        print("Файл не найден")
except PermissionError:
        print("Нет прав доступа к файлу")
except UnicodeDecodeError:
        print("Не удается декодировать файл source.txt")
except Exception as e:
        print(e)
finally:
    source_file.close()
    descination_file.close()

```

```python
#Задание 2. Подсчёт стоимости заказа из файла
try:
    source_file = open('prices.txt','r',encoding='utf-8')
    content = source_file.readlines()
    source_file.close()
    
    sum = 0
    for i in content:
        info = i.strip().split()
        sum += float(info[1]) * float(info[2])
    
    print(sum)

except FileNotFoundError:
        print("Файл не найден")
except PermissionError:
        print("Нет прав доступа к файлу")
except UnicodeDecodeError:
        print("Не удается декодировать файл source.txt")
except Exception as e:
        print(e)
except ValueError as ve:
        print("Ошибка преобразования данных в строке")
except IndexError:
        print("Недостаточно данных в строке")
finally:
    source_file.close()
    
```

```python
#Задание 3. Подсчёт количества слов в файле
try:
    import re
    
    source_file = open('text_file.txt','r',encoding='utf-8')
    content = source_file.readlines()
    source_file.close()
    
    sum = 0
    
    for i in content:
        words = re.findall(r'([A-Za-zА-Яа-яёЁ0-9]+(-[A-Za-zА-Яа-яёЁ0-9]+)?)', i.strip())
        sum += len(words)
    
    print(sum)
except FileNotFoundError:
        print("Файл не найден")
except PermissionError:
        print("Нет прав доступа к файлу")
except UnicodeDecodeError:
        print("Не удается декодировать файл source.txt")
except Exception as e:
        print(e)

finally:
    source_file.close()
```

```python
#Задание 4. Копирование уникального содержимого одного файла в другой
try:
    source_file = open('input.txt','r',encoding='utf-8')
    content = source_file.readlines()
    
    uniq = set(content)
    descination_file = open('unique_output.txt','w',encoding='utf-8')
    
    for i in uniq:
        descination_file.write(i)

except FileNotFoundError:
        print("Файл не найден")
except PermissionError:
        print("Нет прав доступа к файлу")
except UnicodeDecodeError:
        print("Не удается декодировать файл source.txt")
except Exception as e:
        print(e)
finally:
    source_file.close()
    descination_file.close()
```
